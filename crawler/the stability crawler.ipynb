```python
import requests,time,random
from lxml import etree
from urllib import parse
from selenium import webdriver 
import pandas as pd 
import numpy as np
import time
from selenium.webdriver.common.by import By
import random
```

```python
## data and url preparation
ls= pd.read_csv('stability.csv',header=None)
# print(ls)
ls.dropna()
ls=np.array(ls)
query_list=ls.tolist()
url_base='https://www.webofscience.com/wos/alldb/basic-search'
```

```python
# query_list[1746] we have 1747 records 
def split_list(listTemp, step):
    list_result=[]
    for i in range(0, len(listTemp), step):
        list_result.append(listTemp[i: i + step])
    return list_result
```

```python
splitted=split_list(query_list,20)
```

```python
#query_list
options = webdriver.chrome.options.Options()
options.add_argument("--auto-open-devtools-for-tabs")
```

```python
Coutry_list=[]
Facility_list=[]
Journal_list = []
ImpactFactor_list=[]
Index_date_list=[]
Cited_time_list=[]
```

```python
def crawler_by_batch(query_list):
    C_list=[]
    F_list=[]
    J_list = []
    IF_list=[]
    ID_list=[]
    CT_list=[]
    driver = webdriver.Chrome(options=options)
    i=1
    driver.maximize_window()
    for query_text in query_list:
        try:
            ###enter main page
            test=query_text[0]
            try:
                driver.get(url_base)
            except:
                print('link to Webofscience failed')
            if i==1:
                time.sleep(random.uniform(5,15))
            else:
                time.sleep(random.uniform(1.5,2.5))

            ###clear the welcome pages for the first record
            if i==1:
                ###close "coockies" page
                try:
                    driver.find_element(By.XPATH,'//*[@id="onetrust-accept-btn-handler"]').click()
                    time.sleep(1)
                except:
                    print('close coockie skipped')
                ###close "welcome" page
                try:
                    driver.find_element(By.XPATH,'//*[@id="pendo-close-guide-5600f670"]').click()
                    time.sleep(1)
                except:
                    print('close welcome and tutorial skipped')
                ###close "tutorial" page
                try:
                    driver.find_element(By.XPATH,'//*[@id="pendo-close-guide-4d6d27fa"]').click()
                    time.sleep(1)
                except:
                    print('close tutorial skipped')
                ###some times coockies need to be closed again
                try:
                    driver.find_element(By.XPATH,'//*[@id="onetrust-accept-btn-handler"]').click()
                    time.sleep(1)
                except:
                    print('close coockie skipped')
                ###some times welcome need to be closed again
                try:
                    driver.find_element(By.XPATH,'//*[@id="pendo-close-guide-5600f670"]').click()
                    time.sleep(1)
                except:
                    print('close welcome and tutorial skipped')
                ###some times tutorial need to be closed again
                try:
                    driver.find_element(By.XPATH,'//*[@id="pendo-close-guide-4d6d27fa"]').click()
                    time.sleep(1)
                except:
                    print('close tutorial skipped')
            else:
                print('no need for closing welcom pages')
            ###close the survey (before search)
            try:
                driver.find_element(By.XPATH,'//*[@id="pendo-button-54e2fb7b"]').click()
                time.sleep(random.uniform(0.4,0.6))
            except:
                print('no survey need to close')
            ###some times coockies need to be closed again
            try:
                driver.find_element(By.XPATH,'//*[@id="onetrust-accept-btn-handler"]').click()
                time.sleep(1)
            except:
                print('close coockie skipped')
            ###some times welcome need to be closed again
            try:
                driver.find_element(By.XPATH,'//*[@id="pendo-close-guide-5600f670"]').click()
                time.sleep(1)
            except:
                print('close welcome and tutorial skipped')
            ###some times tutorial need to be closed again
            try:
                driver.find_element(By.XPATH,'//*[@id="pendo-close-guide-4d6d27fa"]').click()
                time.sleep(1)
            except:
                print('close tutorial skipped')
            ###close the survey (before search 2)
            try:
                driver.find_element(By.XPATH,'//*[@id="pendo-button-54e2fb7b"]').click()
                time.sleep(random.uniform(0.4,0.6))
            except:
                print('no survey need to close')
            ###clear the search colunm
            try:
                driver.find_element(By.XPATH,'//*[@id="snSearchType"]/div[3]/button[1]/span[1]').click()
            except:
                print('clear input colunmm failed')

            ###choose to search by DOI
            try:
                driver.find_element(By.XPATH,'//*[@id="snSearchType"]/div[1]/app-search-row/div/div[1]/app-select-search-field/wos-select/button').click()
                driver.find_element(By.XPATH,'//*[@id="global-select"]/div/div[2]/div[11]').click()
                time.sleep(random.uniform(1.5,2.5))
            except:
                print('choose DOI failed')
            ###close the survey (after search)
            try:
                driver.find_element(By.XPATH,'//*[@id="pendo-close-guide-5600f670"]').click()
                time.sleep(random.uniform(0.5,0.6))
            except:
                print('no survey need to close')
            ###clear the column,input of DOI number and search
            try:
                if i==1:
                    driver.find_element(By.XPATH,'//*[@id="mat-input-0"]').send_keys('%s\n' % test)
                    time.sleep(random.uniform(8.5,9.5))
                    driver.find_element(By.XPATH,'//*[@id="snSearchType"]/div[4]/button[2]/span[1]').click()
                    time.sleep(random.uniform(2.5,3.5))
                else:
                    driver.find_element(By.XPATH,'//*[@id="mat-input-0"]').send_keys('%s\n' % test)
                    time.sleep(random.uniform(8.5,9.5))
                    try:
                        driver.find_element(By.XPATH,'//*[@id="snSearchType"]/div[4]/button[2]/span[1]').click()
                        time.sleep(random.uniform(2.5,3.5))
                    except:
                        print('no need to click the search bottom”')
            except:
                print("search failed")
            ###close the survey2 (after search)
            try:
                driver.find_element(By.XPATH,'//*[@id="pendo-button-54e2fb7b"]').click()
                time.sleep(random.uniform(0.5,0.6))
            except:
                print('no survey need to close')
            ###close the survey (after search)
            try:
                driver.find_element(By.XPATH,'//*[@id="pendo-close-guide-5600f670"]').click()
                time.sleep(random.uniform(0.5,0.6))
            except:
                print('no survey need to close')
            ###enter the detailed page
            try:
                driver.find_element(By.XPATH,'/html/body/app-wos/main/div/div/div[2]/div/div/div[2]/app-input-route/app-base-summary-component/div/div[2]/app-records-list/app-record/div/div/div[2]/div[1]/app-summary-title/h3/a').click()
                time.sleep(random.uniform(1.5,2.5))
            except:
                print("enter detail page failed")

            ###gather information needed
            try:
                try:
                    indexed_date=driver.find_element(By.XPATH,'//*[@id="FullRTa-indexedDate"]').text
                except:
                    indexed_date='NaN'
                try:
                    cited_times=driver.find_element(By.XPATH,'//*[@id="FullRRPTa-citationsLabelPlural-ALLDB"]').text
                    cited_times=int(cited_times)
                except:
                    cited_times=0
                try:
                    impact_factor=driver.find_element(By.XPATH,'//*[@id="snJournalData"]/div[2]/div[1]/div[1]/app-jcr-sidenav/mat-sidenav-container/mat-sidenav-content/span/button/span[1]/span').text
                    impact_factor=float(impact_factor)
                except:
                    impact_factor=0
                try:
                    journal_name=driver.find_element(By.XPATH,'//*[@id="snJournalData"]/div[1]/app-jcr-sidenav/mat-sidenav-container/mat-sidenav-content/span/a').text
                except:
                    journal_name='NaN'
                try:
                    address=driver.find_element(By.XPATH,'//*[@id="address_1"]/span[2]').text
                except:
                    address='NaN'
                try:
                    Facility=address.split(', ')[0]
                    Country=address.split(', ')[-1]
                except:
                    Facility='NaN'
                    Country='NaN'
                C_list.append(Country)
                F_list.append(Facility)
                J_list.append(journal_name)
                IF_list.append(impact_factor)
                ID_list.append(indexed_date)
                CT_list.append(cited_times)
            except:
                print("gather information failed")
        except:
            print('initial step error.')
        i+=1
        print(i)
        if(i%10==1):
            time.sleep(random.uniform(15,30))
    driver.close()
    print('J_list')
    return C_list,F_list,J_list,IF_list,ID_list,CT_list
```

```python
for each_sub_batch in splitted:
    Coutry_list_batch,Facility_list_batch,Journal_list_batch,IF_list_batch,Index_date_list_batch,Cited_time_list_batch=crawler_by_batch(each_sub_batch)
    #update the list
    Coutry_list=Coutry_list+Coutry_list_batch
    Facility_list=Facility_list+Facility_list_batch
    Journal_list=Journal_list+Journal_list_batch
    ImpactFactor_list=ImpactFactor_list+IF_list_batch
    Index_date_list=Index_date_list+Index_date_list_batch
    Cited_time_list=Cited_time_list+Cited_time_list_batch
```

```python
result=pd.DataFrame([Coutry_list,Facility_list,Journal_list, ImpactFactor_list, Index_date_list, Cited_time_list]).T
new_col = ['Country', 'Facility', 'Journal Name', 'IF', 'Indexed Date','Cited Time']
result.columns = new_col
result
```
